# Results analysis

## Metric 

$Border\;Error=\frac{Area((Automatic\;detection)\oplus(Ground\;truth))}{Area(Ground\;truth)}100\%$

```python
import ProjectLIB as PL
import matplotlib.pyplot as plt
import os.path
import numpy as np
from PIL import Image
```

## Border Error Melanoma

```python
mel_nums = ['030', '046', '049', '140', '142', '143', '145', '146', '150', '151']
folder = 'melanoma'
imgname = 'ISIC_0000'

BordErr_mel = np.zeros(10)

n = 0
for i in mel_nums:

    path_D = os.path.join("dataset", 'melanoma', "SEG_"+imgname+i+'.jpg')
    path_GT = os.path.join('dataset', 'melanoma', imgname+i+'_Segmentation.png')
    
    seg_D = Image.open(path_D)
    seg_D = PL.RGBtoGRAYnumpy(seg_D)
    seg_GT = Image.open(path_GT)
    seg_GT = PL.RGBtoGRAYnumpy(seg_GT)
    
    BordErr_mel[n] = 100*(np.count_nonzero((seg_GT^seg_D))/ np.count_nonzero(seg_GT))
    
    print('Border Error of the segmentation for image '+imgname+i+' = ', BordErr_mel[n])
    
    n=n+1

```

## Masks superposition Melanoma

```python
for i in mel_nums:

    path_D = os.path.join("dataset", 'melanoma', "SEG_"+imgname+i+'.jpg')
    path_GT = os.path.join('dataset', 'melanoma', imgname+i+'_Segmentation.png')
    
    seg_D = Image.open(path_D)
    seg_D = PL.RGBtoGRAYnumpy(seg_D)
    seg_GT = Image.open(path_GT)
    seg_GT = PL.RGBtoGRAYnumpy(seg_GT)
    
    plt.figure(figsize=(15,8))
    plt.imshow(seg_GT^seg_D, cmap='gray')
    plt.title('Blackhat')
    plt.show()
```

### Dice coeficient

$Border\;Error=\frac{2(Area(Automatic\;detection)\cap Area(Ground\;truth)))}{Area(Ground\;truth)+Area(Automatic\;detection)}100\%$

```python
mel_nums = ['030', '046', '049', '140', '142', '143', '145', '146', '150', '151']
folder = 'melanoma'
imgname = 'ISIC_0000'

Dice_mel = np.zeros(10)

n = 0
for i in mel_nums:

    path_D = os.path.join("dataset", 'melanoma', "SEG_"+imgname+i+'.jpg')
    path_GT = os.path.join('dataset', 'melanoma', imgname+i+'_Segmentation.png')
    
    seg_D = Image.open(path_D)
    seg_D = PL.RGBtoGRAYnumpy(seg_D)
    seg_GT = Image.open(path_GT)
    seg_GT = PL.RGBtoGRAYnumpy(seg_GT)
    
    Dice_mel[n] = 100*(2*np.count_nonzero((seg_GT&seg_D)))/(np.count_nonzero(seg_GT) +  np.count_nonzero(seg_D))
    
    print('Dice coefficient of the segmentation for image '+imgname+i+' = ', Dice_mel[n])
    
    n=n+1

```
